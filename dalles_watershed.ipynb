# Notebook to create figure displaying the Dalles Dam watershed

```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import geoviews as gv
import geoviews.feature as gf
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.img_tiles as cimgt
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

gv.extension('bokeh', 'matplotlib')
```

```python
%matplotlib inline
```

```python
datafile = './waterhackweek_dalles_salmonfalls/aggUH_TDA.nc'
ds = xr.open_dataset(datafile)
```

```python
ds
```

```python
#Make a plot of the Dalles Dam Watershed

LINEWIDTH=1.25
EDGECOLOR = 'gray'

mask = ds.fraction
mask = mask.where(mask == 1 )

stamen_terrain = cimgt.Stamen('terrain-background')

states_provinces = cfeature.NaturalEarthFeature(
    category='cultural',
    name='admin_1_states_provinces_lines',
    scale='50m',
    facecolor='none',
    edgecolor=EDGECOLOR,
    linewidth=LINEWIDTH)




fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection=stamen_terrain.crs)
ax.add_image(stamen_terrain, 5)

              
mask.plot.contourf(ax=ax, transform=ccrs.PlateCarree(), alpha=0.7, antialiased=True, add_colorbar=False, add_labels=True)
ax.plot(-121.08, 45.68, "^", markersize=10, color='black', transform=ccrs.Geodetic())
ax.text(-122.75, 45.92,'Dalles Dam', fontsize=12, fontweight='bold', family='serif', transform=ccrs.Geodetic())


ax.add_feature(states_provinces)
ax.add_feature(cfeature.BORDERS.with_scale('50m'),linewidth=LINEWIDTH,edgecolor=EDGECOLOR)
ax.add_feature(cfeature.COASTLINE.with_scale('50m'),linewidth=LINEWIDTH,edgecolor='black')
ax.add_feature(cfeature.RIVERS.with_scale('50m'),linewidth=LINEWIDTH,edgecolor='royalblue')



ax.set_xticks(np.linspace(-130,-105,6), crs=ccrs.PlateCarree())
ax.set_yticks(np.linspace(40,55,4), crs=ccrs.PlateCarree())
lon_formatter = LongitudeFormatter(zero_direction_label=True)
lat_formatter = LatitudeFormatter()
ax.xaxis.set_major_formatter(lon_formatter)
ax.yaxis.set_major_formatter(lat_formatter)

ax.set_extent([-130, -105, 40.5, 53])


```
